```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/COMPETENCIAS /NASA - COVID - MIT_critical_data/data/data_space_pre_processed.csv")
df.info()
```

```python
import copy
columns = ['LAT', 'LON','YEAR', 'MO', 'DY', 'T2M', 'T2MDEW', 'TS', 'PRECTOTCORR', 'PS']
covariant = df[columns].copy()
covariant
```

```python
# Set Data type as int
date_columns = ['YEAR', 'MO', 'DY']
covariant[date_columns] = df[date_columns].astype('int')
covariant[date_columns].dtypes

def merge_dates(x):
    #print(x)
    #print("----")
    return str(x[0]) + '-' + str(x[1]) + '-' + str(x[2])

# Get Date in Day - Month - Year format
covariant['Date'] = covariant[date_columns].apply(merge_dates, axis = 1)
covariant['Date'] = pd.to_datetime(covariant['Date'])
# Drop Day, Month, Year
covariant.drop(columns = date_columns, axis=1, inplace=True)
covariant.head(3)
```

```python
covariant.drop('LAT', axis=1, inplace=True)
covariant.drop('LON', axis=1, inplace=True)
covariant
```

```python
# Download Data as csv
covariant.to_csv('spatial_data_merge_by_date.csv')
```
