```python
import json
import geopandas as gpd
import pandas as pd
import geojson
```

```python
departments = "../Data/Shapefiles/departments_and_risk.geosjon"
# Opening JSON file
f = open(departments,)
 
# returns JSON object as
# a dictionary
data = json.load(f)
```

```python
data
```

```python
df_places = gpd.read_file(departments)
df_places.NOMBRE_DPT[df_places.NOMBRE_DPT == 'archipielagodesanandres,providenciaysantacatalina'] = 'sanandres'
df_places.NOMBRE_DPT = df_places.NOMBRE_DPT.str.strip()
df_places.head()
```

```python
mutations = pd.read_csv('../Data/Covid19_alpha_preprocessed.csv')
mutations.drop(columns=['Unnamed: 0'], inplace=True)
mutations.index = mutations.Date
mutations = mutations.T
mutations = mutations.iloc[1:].reset_index()
mutations.rename(columns={'index':'NOMBRE_DPT'}, inplace=True)
mutations.NOMBRE_DPT = mutations.NOMBRE_DPT.str.strip()
mutations
```

```python
mutations.to_csv('../Data/Covid19_alpha_preprocessed_final.csv')
```

```python
mutations = mutations.merge(df_places, how='inner', on=['NOMBRE_DPT'])
mutations = mutations.reset_index()
mutations
```

### Population

```python
population = pd.read_csv('../Data/Census_data.csv')[['total_population', 'Dpto']]
population.rename(columns={'Dpto':'NOMBRE_DPT'}, inplace=True)
population
```

```python
mutations = mutations.merge(population, on=['NOMBRE_DPT'],how='inner')
```

```python
#mutations["geometry"]
```

```python
poly = {
        'type': 'Polygon',
        'coordinates': mutations["geometry"]
}
```

```python
data = {
    'type': 'features',
    'id':mutations['index'],
    'properies':{        
        'name':mutations['NOMBRE_DPT'],
        'risk':mutations['risk']
    }
}
```

```python
df = {
    'type': 'FeatureCollection',
    'features': data,
    'geometry': poly 

}
```

```python
gdf = gpd.GeoDataFrame(mutations)
gdf
```

```python
with open('../Data/data_alpha.geojson','w') as f:
    geojson.dump(gdf, f)
```
